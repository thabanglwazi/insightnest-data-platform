Understanding these price patterns is vital for both consumers and producers to make informed decisions about purchasing, stocking, and policy planning.

This analysis can provide useful insights for:

Consumers, to plan purchases and budget more effectively

Retailers and suppliers, to optimize stock levels and pricing strategies

Policy makers, to monitor food security and inflation triggers

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Import table
df = pd.read_csv("C:/Users/mathe/OneDrive/Desktop/prices.csv")

# View sample data
print(df.head())
```

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

```

```python
#Convert Date column to datetime
df['Price Dates'] = pd.to_datetime(df['Price Dates'], dayfirst=True, errors='coerce')

# Display the first few rows to confirm
print(df[['Price Dates']].head())
```

```python
print(df.columns)

```

```python
#Basic statistics (mean & standard deviation)
mean_prices = df.mean().round(2)
std_prices = df.std().round(2)

print("\nAverage Prices (ZAR/unit):")
print(mean_prices)

print("\nPrice Volatility (Standard Deviation):")
print(std_prices)
```

```python
# Heatmap of correlation between vegetables
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Vegetable Prices")
plt.show()
```
